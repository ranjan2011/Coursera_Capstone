```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(req.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
neighborhood=pd.DataFrame(df[0])
```

```python
neighborhood.head()
```

```python
column_names = ['PostalCode', 'Borough', 'Neighborhood']
Toronto_df = pd.DataFrame(columns=column_names)
```

```python
neighborhood.values.tolist()[0][0][0:3]
```

```python
len(neighborhood.values.tolist()[0])
```

```python
firstlist=[]
secondlist=[]
for i in range(20):
    for j in range(9):
        val=neighborhood.values.tolist()[i][j][0:3]
        val2=neighborhood.values.tolist()[i][j][3:]
        firstlist.append(val)
        secondlist.append(val2)
```

```python
Toronto_df
```

```python
Toronto_df["PostalCode"]=firstlist
Toronto_df["Borough"]=secondlist
```

```python
Toronto_df.shape
```

```python
Toronto_df.drop(Toronto_df.index[Toronto_df['Borough'] =="Not assigned"], inplace = True)
```

```python
Toronto_df["Borough"].tolist
```

```python
Toronto_df=Toronto_df.reset_index(drop=True)
```

```python
Toronto_df
```

```python
thirdlist=[]
fourlist=[]
for i in range(103):
    thirdlist.append(Toronto_df["Borough"].tolist()[i].split("(")[0])
    fourlist.append(Toronto_df["Borough"].tolist()[i].split("(")[1])
    
```

```python
Toronto_df["Borough"]=thirdlist
Toronto_df["Neighborhood"]=fourlist
```
